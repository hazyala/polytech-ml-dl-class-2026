### 정육면체 주사위를 10번, 1000번, 10000번 굴려서 나온 숫자들의 빈도수를 계산하여 출력하는 프로그램

```python
import numpy as np
import matplotlib.pyplot as plt

few_rolls = np.random.randint(1,7,size=10)
many_rolls = np.random.randint(1,7,size=1000)
max_rolls = np.random.randint(1,7,size=10000)

few_counts = np.histogram(few_rolls, bins=np.arange(0.5,7.5))[0]
many_counts = np.histogram(many_rolls, bins=np.arange(0.5,7.5))[0]
max_counts = np.histogram(max_rolls, bins=np.arange(0.5,7.5))[0]

fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(12,3))
ax1.bar(np.arange(1,7),few_counts)
ax2.bar(np.arange(1,7),many_counts)
ax3.bar(np.arange(1,7),max_counts)

plt.show()
```

### 확률 분포

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

b = ss.distributions.binom

for flips in [5, 10, 20, 40, 80]:
    successes = np.arange(flips)
    out_distribution = b.pmf(successes, flips, 0.5)
    plt.hist(successes, flips, weights=out_distribution)
    
plt.xlim(0, 55)

plt.show()
```

### 선형결합, 가중 합, 점곱

```python
quantity = [2,12,3]
costs = [12.5,0.5,1.75]
partical_cost=[]
for q,c in zip(quantity,costs):
    partical_cost.append(q*c)
print(sum(partical_cost))
```

```python
import numpy as np
quantity = [2,12,3]
costs = [12.5,0.5,1.75]
print(np.array(quantity)*(np.array(costs)))
print(sum(np.array(quantity)*np.array(costs)))
```

```python
import numpy as np

quantity = np.array([2,12,3])
costs = np.array([12.5,0.5,1.75])
print(quantity.dot(costs))
print(np.dot(quantity, costs))
print(quantity @ costs)
```

### 단순평균, 가중 평균

```python
import numpy as np
values = np.array([10.0,20.0,30.0])
weights = np.full_like(values, 1/3)

print("weights :", weights)
print("via mean :", np.mean(values))
print("via weights and dot :", np.dot(weights, values))
```

```python
import numpy as np

values = np.array([10.0,20.0,30.0])
weights = np.array([0.5, 0.25, 0.25])
print("weights :", weights)
print("via weights and dot :", np.dot(weights, values))
```
